```
salidas= ['OUTPUT','aroma','Sabor','residual','Acidez','Cuerpo','Balance','Pcatador']
```

```
matrix.drop(labels=salidas).loc[(matrix['OUTPUT'] >= 0.5) | (matrix['OUTPUT'] <= -0.5) ]['OUTPUT'].sort_values(ascending=False)
```

```
matrix.drop(labels=salidas).loc[(matrix['aroma'] >= 0.5) | (matrix['aroma'] <= -0.5) ]['aroma'].sort_values(ascending=False)
```

```
matrix.drop(labels=salidas).loc[(matrix['Cuerpo'] >= 0.5) | (matrix['Cuerpo'] <= -0.5) ]['Cuerpo'].sort_values(ascending=False)
```

```
matrix.drop(labels=salidas).loc[(matrix['Balance'] >= 0.5) | (matrix['Balance'] <= -0.5) ]['Balance'].sort_values(ascending=False)
```

```
matrix.drop(labels=salidas).loc[(matrix['Pcatador'] >= 0.5) | (matrix['Pcatador'] <= -0.5) ]['Pcatador'].sort_values(ascending=False)
```

```
interesantes = {}
for lab in salidas:
  interesantes[lab] = list(matrix.drop(labels=salidas).loc[(matrix[lab] >= 0.5) | (matrix[lab] <= -0.5) ][lab].sort_values(ascending=False).index)
```

```
interesantes
```

```
df_list={}
for k in interesantes.keys():
  df_list["df_"+k] = Coffee[interesantes[k]]
```

```
for k in df_list.keys():
  df_list[k].to_csv(k+".csv")
```
